```python
"""
To classify the input skin into one of the 6 skin tones
"""
import cv2
import numpy as np
import pandas as pd
import sklearn
from skin_detection import skin_detection
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("skin_tone_dataset.csv")
X = df.iloc[:, [1, 2, 3]].values
y = df.iloc[:, 0].values
```

```python
classifier = KNeighborsClassifier(n_neighbors=6, metric='minkowski', p=2)
classifier.fit(X, y)
```

```python
# Use skin_detection function to get RGB values for the test image
X_test = skin_detection(r"test images/9.jpg")
```

```python
X_test = X_test.reshape(1, -1)
y_pred = classifier.predict(X_test)
print(y_pred)
```
